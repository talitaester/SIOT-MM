```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from tqdm.notebook import trange, tqdm
from scipy.spatial import distance

#trace = pd.read_csv("D:/Users/Talit/Downloads/SIOT-MM-implementations-main/SIOTMM/logs/aSIOTmm_30-05-2023_21.13.17.265/pos_trace.csv", index_col=False)
#trace = trace[(trace[' time'] % 60) == 0]
##trace = trace.rename(columns={' y': 'y', ' x': 'x', ' time' : 'time'})
#print(trace.head())
import pandas as pd
import numpy as np
import math

def euclidean_distance(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance

# Carregar os dados do dataframe (substitua "df" pelo nome do seu dataframe)
df = pd.read_csv("D:/Users/Talit/Downloads/SIOT-MM-implementations-main/SIOTMM/logs/aSIOTmm_30-05-2023_21.13.17.265/pos_trace.csv", index_col=False)

stay_locations = []
geolocations = []
travel_time = []
visit_time = []

for obj_id in df["id"].unique():
    obj_data = df[df["id"] == obj_id].reset_index(drop=True)
    
    stay_location_threshold = 20
    stay_location = []
    prev_x, prev_y = obj_data.loc[0, " x"], obj_data.loc[0, " y"]
    for i in range(1, obj_data.shape[0]):
        curr_x, curr_y = obj_data.loc[i, " x"], obj_data.loc[i, " y"]
        distance = euclidean_distance(prev_x, prev_y, curr_x, curr_y)
        if distance >= stay_location_threshold:
            stay_location.append((prev_x, prev_y))
        prev_x, prev_y = curr_x, curr_y
    stay_locations.append(stay_location)
    
    # calculo das geolocations
    geolocation_threshold = 150 
    geolocation = []
    prev_time = obj_data.loc[0, " time"]
    for i in range(1, obj_data.shape[0]):
        curr_time = obj_data.loc[i, " time"]
        time_diff = curr_time - prev_time
        if time_diff >= geolocation_threshold:
            geolocation.append((obj_data.loc[i, " x"], obj_data.loc[i, " y"]))
        prev_time = curr_time
    geolocations.append(geolocation)
    
    # calculo do travel time
    travel_time.append(obj_data[" time"].iloc[-1] - obj_data[" time"].iloc[0])
    
    # calculo do visit time
    visit_time.append(sum(obj_data[" time"].diff() >= geolocation_threshold))

# exibir os resultados
for i in range(len(stay_locations)):
    print("Objeto:", i+1)
    print("Stay Locations:", stay_locations[i])
    print("Geolocations:", geolocations[i])
    print("Travel Time:", travel_time[i])
    print("Visit Time:", visit_time[i])
    print()
```
