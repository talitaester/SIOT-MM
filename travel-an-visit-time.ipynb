```python
import math

def euclidean_distance(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance

list_concat = []
glX = []
glY = []
slX = []
slY = []
moment = []
time_final = []
visit_time =[]
travel_time = []
travel_time_final = []
nodes = trace['id'].unique()

for id in nodes:
    
    temp_trace = trace[trace['id'] == id]
    xList = temp_trace[' x'].to_list()
    yList = temp_trace[' y'].to_list()
    rows = len(xList) - 1
    moment_list = temp_trace[' time'].to_list() 
    h = 0
    for y in range(1, len(yList)-1):
        distance = euclidean_distance(xList[h], yList[h], xList[y], yList[y])
        delta_space = 20
        h = y
        if distance > delta_space:
            slX.append(xList[y])
            slY.append(yList[y])
            time_final.append(moment_list[y])
    ##geo-location
    p = 0
    if len(time_final) > 0:
        for x in range(1, len(time_final)-1):
            dif = time_final[x] - time_final[p]
            delta_time = 100
            #counting visit time
            if dif > delta_time:  
                visit_time.append(dif)
                glX.append(slX[x])
                glY.append(slY[x])
                moment.append(time_final[x])
                p = x
                #counting travel time 
                #w = len(travel_time) -1
                #if len(travel_time)  >= 2:
                #    dif2 = travel_time[w] - travel_time[0]
                #    travel_time_final.append(dif2)
#
                #travel_time.clear() 
            #else: 
                #dif2 = travel_time[w] - travel_time[0]
                #travel_time.append(time_final[x])


    print(id)
    for_plot = visit_time
    #print('visit time: ', visit_time)
    #print('travel time: ', travel_time_final)
    travel_time.clear()
    travel_time_final.clear()
    #visit_time.clear()
    time_final.clear()
    glX.clear()
    glY.clear()
    slX.clear()
    slY.clear()
    travel_time.clear()
```
